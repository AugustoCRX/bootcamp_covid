```python
def LangDetector(text_nlp):
    #Carrega modelo de leitura de detecção de linguagem
    nlp = spacy.load('es_core_news_sm')
    #Cria um pipeline com a função definida
    nlp.add_pipe('language_detector',last=True)
    text = text_nlp
    doc = nlp(text)
    return doc._.language['language'],doc._.language['score']
```

```python
def read_json_file(file):
    with open(file, "r", encoding='utf8') as r:
        response = r.read()
        response = response.replace('\n', '')
        response = response.replace('}{', '},{')
        response = "[" + response + "]"
        return json.loads(response)
```

```python
#função de diminuição de dimensionalidade de listas
def flatten_list(_2d_list):
    flat_list = []
    for element in _2d_list:
        if type(element) is list:
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list
```

```python
caminho_MX = r"DATA"
list_MX = glob.glob(os.path.join(caminho_MX, '*/*'))

# extraindo json e criando lista
dict_MX = list()
for i in range(len(list_MX)):
    dict_MX.append(read_json_file(list_MX[i]))
dict_MX = flatten_list(dict_MX)
```

```python
#Extração do "place_id" para coleta de dados geográficos
data_extraction = []
for i in range(len(dict_MX)):
    try:
        for j in range(len(dict_MX[i]['data'])):
            place_id = dict_MX[i]['data'][j]['geo']['place_id']
            data_extraction.append(place_id) 
    except:
        continue  
data_extraction = list(set(data_extraction))
```

```python
path = os.listdir('GEO')
dict_ES_GEO = list()
for i in path:
    with open(f'GEO/{i}', 'r', encoding='utf8') as f:
        dict_ES_GEO.append(json.load(f))
for i in range(len(dict_ES_GEO)):
    try:
        if dict_ES_GEO[i]['id'] in data_extraction:
            data_extraction.remove(dict_ES_GEO[i]['id'])
    except:
        continue  
```

```python
path = os.listdir('GEO')
dict_ES_GEO = list()
for i in path:
    with open(f'GEO/{i}', 'r', encoding='utf8') as f:
        dict_ES_GEO.append(json.load(f))
```

```python
# A estrutura da lista funciona com o seguinte padrão:
# O primeiro objeto (dict_AR[i]) é o tweet de várias pessoas que falaram sobre o assunto covid naquele dia
# Dentro do objeto citado anteriormente podemos acessar as seguintes informações:
######### data #########
# Inclui os dados do tweet em si 
######### includes #########
# Inclui os dados do usuário que tweetou
######### meta #########
# Inclui informações sobre a pesquisa
# Com isso, as informações que serão coletadas serão:
# - Texto
# - Data
# - Quantidade de Retweet
# - Quantidade de likes
# - Grau inicial de polaridade (Utilizando a biblioteca TextBlob)
# O grau inicial será levado como consideração como baseline inicial

# all_text = []
# for i in range(len(dict_AR)):
#     try:
#         for j in range(len(dict_AR[i]['data'])):
#             text = dict_AR[i]['data'][j]['text']
#             for tweet in [dict_AR[i]['data'][j]['text']]:
#                 analysis = tb(tweet)
#                 polarity = analysis.sentiment.polarity
#             retweet = dict_AR[i]['data'][j]['public_metrics']['retweet_count']
#             like = dict_AR[i]['data'][j]['public_metrics']['like_count']
#             date = dict_AR[i]['data'][j]['created_at'].split('T')[0]
#             all_text.append(dict(text = text,
#                                 score = polarity,
#                                 retweet = retweet,
#                                 like = like,
#                                 date = date
#                                 )) 
#     except:
#         continue  

all_text = []
for i in range(len(dict_MX)):
    try:
        for j in range(len(dict_MX[i]['data'])):
            text = dict_MX[i]['data'][j]['text']
            for tweet in [dict_MX[i]['data'][j]['text']]:
                analysis = tb(tweet)
                polarity = analysis.sentiment.polarity
            retweet = dict_MX[i]['data'][j]['public_metrics']['retweet_count']
            like = dict_MX[i]['data'][j]['public_metrics']['like_count']
            date = dict_MX[i]['data'][j]['created_at'].split('T')[0]
            for z in range(len(dict_ES_GEO)):
                if dict_ES_GEO[z]['id'] == dict_MX[i]['data'][j]['geo']['place_id']:
                    lat = dict_ES_GEO[z]['centroid'][1]
                    long = dict_ES_GEO[z]['centroid'][0]
            all_text.append(dict(text = text,
                                score = polarity,
                                retweet = retweet,
                                like = like,
                                date = date,
                                lat = lat,
                                long = long
                                )) 
    except:
        continue  
```

```python
MX_df = pd.DataFrame(all_text)
MX_df.to_csv('MX_data1.csv')
```
